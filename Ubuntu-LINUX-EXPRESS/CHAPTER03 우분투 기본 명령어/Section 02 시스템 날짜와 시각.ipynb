# 1. 현재 시각과 날짜 출력

운영체제의 시각은 두가지 형태 제공

- 1. 마더보드(메인보드)에서 배터리 전력으로 구동되는 하드웨어 시계
     RTC(Real-Time Clock)
     - 운영체제 간섭 받지 않은 독립적인 형태로 동작


- 2. 커널에 의해 관리되는 소프트웨어 시계(System Clock)
     - 커널에 의해 관리

## 현재 시각과 날짜 출력 : date

```shell
    date
```


## 시간 동기화 : timedatectl

하드웨어 시계와 소프트웨어 시계, UTC,타임존, 시간 동기화 등을 출력하고자 할 때
```shell
timedatectl
```

# 2. 연간 달력 출력

오늘 날짜 해당하는 월의 달력을 출력  : cal 명령 사용

## 오늘의 날짜가 속한 달 출력 : cal
```shell
cal
```


## 달 출력 : cal[옵션]

```shell
cal 2024
```
