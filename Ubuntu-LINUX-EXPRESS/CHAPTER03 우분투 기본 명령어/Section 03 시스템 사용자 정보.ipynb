# 1. 로그인 네임과 ID 확인

우분투 

 - 실시간 많은 사용자들이 로그인하여 사용
 - 동시에 접속한 사용자들의 로그인 정보와 사용자 계정 ID에 대한 정보 확인 필요


## 사용중인 로그인 네임 : logname

 - 우분투 사용자들이 동시에 접속하여 사용 중인 로그인 네임을 확인하고자 할 때
   logname 명령을 사용

```CLI
    logname
```


## 접속한 사용자의 아이디 : users

 - 우분투에 접속한 사용자들의 ID 정보를 확인하고자 할때 users 명령 사용
 - 현재 접속한 사용자의 아이디를 출력

```CLI
users
```



---

# 2. 사용자 계정 정보

 - 우분투에 로그인 한 모든 사용자의 계정과 터미널, 로그인 시각 등을 확인
 - 우분투를 사용하고 있는 사용자가 누구인지 정보를 확인하는 명령어

## 로그인 한 모든 사용자 계정 : who

```cli
who
```


## 현재 우분투 사용자 확인 : whoami

 - 가상 콘솔을 사용하거나 여러 계정으로 동시에 우분투에 접속하여 로그인 되어 있는 경우
   사용자에 대한 구분이 혼란스러울수 있음 
 - 현재 우분투를 사용하고 있는 사용자가 누구인지 확인하고자 할 때 whoami 명령을 사용

```cli
whoami
```
