우분투 모든 처리과정을 파일 단위로 처리

 - 계층적인 구조의 특성

우분투에서 다루고 있는 파일은 3개의 충족요건을 갖추고 있어야 함

 - 1. 파일 이름
 - 2. I-node
 - 3. 데이터 블록
 


---

# 1. 파일 이름

 - 사용자가 파일을 사용목적에 맞도록 사용
 - 해당 파일에 정확하게 접근하게 접근할 수 있도록 구별할 수 있는 변별력 제공

 - !! window운영체제 파일의 속성에 따른 확장명 개념 사용
     - 우분투 특별한 확장명 제공하지 않음 
     - 우분투 관리 쉽도록 파일을 구분하기 위한 차원에서 사용되는 정도

---

# 2. I-node

I-node
    
    - Index Node 라고 불림
    - 파일을 기술하는 디스크 상에서 데이터 구조를 의미
    - 파일의 데이터 블록이 디스크 상의 어느 주소에 위치하고 있는가에 대한 정보 기록용 사용
    
    - 일반적 파일 생성
         - 1. I-node의 link '0' 위치에 I-node 생성
         - 2. 정보 저장 후 link '1'위치로 이동
         
 - I-node에 저장된 정보
     - 파일의 종류
     - 파일의 소유권 - 사용자(소유자)와 그룹
     - 파일의 액세스 모드
     - 파일의 타임스탬프(파일 갱신일)
     
 - 우분투 파일 생성
     - 1. 저장하기 위한 디스크 내의 공간 먼저 확보
     - 2. 저장될 주소에 대한 정보 기록 하기 위해 
          I-node 각 파티션마다 0부터 시작하는 정수의 형태로 고유의 식별번호를 가짐

---

# 3. 데이터 블록

 - 데이터블록(Data Block) 파일에서 데이터를 저장하는 블록을 의미
 - 일반 파일이나 디렉터리 파일의 데이터가 존재
 - 디스크 장치에 파일을 저장할 때 실제 데이터는 특별한 구분 없이 디스크에 저장
 - 파일1과 파일2가 디스크에 저장될 때 특별히 2개를 따로 구분하지 않고 
   디스크에 차례대로 기록
  
 - 파일 실제 데이터 디스크의 어느 한 곳에 차례대로 쌓여 있음
 - 이부분을 데이터 블록이라고 함
 
 - 파일 구분하지 않기 때문에 구분 부가 정보 필요
 - I-node를 이용 부가 정보를 구별할 수 있도록 편의성 제공 

    - 블록 그룹은 블록들의 모임을 의미
    - os커널이 같은 파일에 속하는 데이터 블록은 블록 그룹에 저장하는 기능을 제공
      파일 단편화 줄일 수 있음
    
    - 파일 시스템의 전체적인 정보 
        - 슈퍼블록(Super Block)과 그룹 디스크립터 테이블(Group Descriptor Table)에 저장
