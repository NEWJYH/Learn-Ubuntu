우분투 관리 모든 데이터 
    
    - 파일(현재 실행중인 프로세스, 일반 파일, 하드웨어 등)로 인식
    
    - 하나의 프로세스는 하나의 작업 디렉터리를 가짐
    
    - 우분투는 3개의 파일 종류(일반, 디렉터리, 특수)파일을 가짐
    
    - 파일을 사용자가 이용할 수 있는 데이터의 실체를 의미
    


### !우분투 시스템이나 응용프로그램도 파일 취급

    - 키보드, 프린터, LAN 카드, 마우스 등도 파일로 다룸
    - 우분투 모든 자원을 파일로 다룸ㅈ

# 1. 일반 파일

 - 일반 파일(Regular File) 데이터를 관리하는데 주로 사용되는 일상적 파일
 - 각종 실행 파일, 텍스트 파일, 이미지 파일, 프로그램 소스 파일 등
 - 실행파일, 이미지 파일  
     - 바이너리 형태로 저장

### 실행파일 확인 
    
    - 유틸리티 프로그램 있어야 함
    - 텍스트 파일 : 문서 편집기
    - 일반 파일은 정규 파일이라고도 함 
      정규 파일은 표준 입출력 시스템의 호출을 통해 참조

---

# 2. 디렉터리 파일


- 디렉터리(Directory) 파일로 취급

- 해당 디렉터리에 저장된 파일이나 하위 디렉터리에 대한 정보를 가짐
- 디스크에 저장 다른 파일을 조작하고 엑세스하는데 필요한 정보를 가짐


### 디렉터리파일 안 엑세스 필요한 정보 가짐
### 전체 파일을 하나의 트리구조로 만들어 관리

---

# 3. 링크 파일

 - window 사용자에게 낯선 느낌
 
 - 편리성을 제공
 - 유닉스 & 리눅스 Link(링크) 개념 도입
     - 여러 개의 파일 이름이 하나의 I-node에 연결되도록 수행
     - 실제 파일이나 디렉터리 혹은 또 다른 링크를 가르킴
     
### 링크 추가
    - 단순하게 디렉터리의 엔트리를 만드는 것
    - 하드 링크와 심볼릭 링크 존재
    - 링크 파일을 생성하기 위해 ln명령을 사용

## 하드 링크(Hard Link)

    - 원본 파일 복사하여 원본 파일과 동일한 내용의 다른 사본파일을 만드는 것
    - 하드 링크에서는 원본 파일과 링크 파일 두개가 서로 다른 파일로 취급
    - 둘 중 하나 삭제 하더라도 나머지 하나는 그대로 남아 있음
    
    
### 하드 링크
     - 원본 파일의 내용이 변경될 경우 복사된 링크 파일의 내용 또한 자동으로 변경
     - 하드 링크는 동일한 I-node를 갖는 파일을 생성하는 것을 의미
     - 파일의 실제 I-node의 정보를 공유 
     - 마치 동일한 파일이 여러 곳에 존재하는 것처럼 보임

## 심볼릭 링크(Symbolic Link)

 - 우분투에서 통상적으로 사용되는 링크 파일
 - 소프트 링크(Soft Link)라고도 함
 - window 운영체제 바로 가기 기능이나 단축 아이콘과 같은 역할을 수행
 
 ### 이미 생성된 심볼릭 링크 파일을 삭제하여도 원본 영향 x
 ### 원본파일이 삭제 될 경우 링크 파일은 깜빡 거림(없다는 것을 알림)

# 4. 특수 파일

 - 우분투 지원하는 파이프, 소켓, 디바이스(주변장치)를 의미
 - window 운영체제 하드디스크(Hard Disk)가 C:, D:, E: 등으로 표시
   우분투에서는 /dev/hda1, .. /dev/hda3과 같이 사용

# 5. 문서 파일과 이진 파일

## 문서 파일
 - 문서 파일(Text File)은 키보드로 입력받은 값만 가지고 있음
 - ASCII 문자로 구성
     - 알파벳 대/소문자, 숫자, 공백, 탭, 제어문자, 문장부호 포함 128개의 코드로 구성
     - 문서를 작성할 때 해당 알파벳을 ASCII문자로 매칭

## 이진 파일
 - 이진 파일(Binary File)은 문자가 아닌 데이터 파일이 들어 있음 
 - 기계어인 '0','1'값으로 구성
 - 프로그램을 코딩하여 컴파일을 수행하는 과정에서 생성되는 파일 의미
 - 실행 파일이라고도 불림
