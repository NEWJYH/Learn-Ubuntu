# 1. 디렉터리 계층 구조 

- 계층적 트리(Tree)구조로 구성
- 사용하고자 하는 목적, 수행되는 명령어의 성격과 내용 및 사용권한 부여 등
  목적에 따라 디렉터리로 구분하여 제공

최상위 root 디렉터리 기반 - 루트 디렉터리 또는 최상위 디렉터리라고 함
 - 일반적 루트 디렉터리 /(슬래시)로 표시
 - 루트 디렉터리 제외 모든 디렉터리 서브 디렉터리 또는 하위 디렉터리라고 부름

## 루트 디렉터리와 서브 디렉터리

루트 디렉터리 
    - 기본적으로 서브 디렉터리를 가짐

      - /는 디렉터리 파일을 의미 
      - @는 심볼릭 링크 파일 의미 

우분투 시스템의 루트 디렉터리와 심볼릭 링크 확인 위해 -F 옵션과 루트를 의미하는 /를 선언

- 존재하는 리스트를 확인

```cli
    ls -F /
```


## 작업 디렉터리

- 사용자가 우분투에 접속하여 현재 사용하고 있는 디렉터리를 의미
- 현재 작업중인 디렉터리는 점(.)으로 표시
- 현재 작업중인 디렉터리의 위치를 상세하게 확인하기 위해서는 pwd 명령을 사용

## 홈 디렉터리

- 각 사용자에게 할당되는 디렉터리
- 사용자 계정을 처음 만들 때 지정해줌
- 많은 사용자 접속하여 사용 서버 시스템 구조 각각 사용자별로 홈 디렉터리를 지정해줘야 함

#### 사용자에게 지정된 홈 디렉터리에 파일과 서브 디렉터리 생성하여 작업 가능
#### 홈 디렉터리의 표시는 틸트(~)를 사용하여 표시
#### ex) guest_01 계정의 홈 디렉터리는 'guest_01 ~'과 같이 표시

# 2. 절대 경로와 상대 경로

경로(Path)는 특정 파일이나 디렉터리의 위치를 나타냄

- 각 경로는 구분자 '/'를 사용하여 구분되도록 표기
- 절대경로(Absolute Path)와 상대경로(Relatvie Path) 두 가지의 형태로 제공

#### 절대 경로 
    - root 디렉터리인 /를 기준으로 제공되는 디렉터리를 의미
    
#### 상대경로
    - 절대 경로인 루트 /를 포함하지 않는 경로
    - 셸의 작업 디렉터리가 상대 경로의 기준
    - 현재 자신이 속해 있는 디렉터리는 . 으로 표현
    - 바로 위 상위 디렉터리는 .. 으로 표현 

## 절대 경로
 - 항상 루트(/) 디렉터리 위치부터 시작
     - 반드시 루트 의미하는 /로 시작해야 함
     - 루트 디렉터리부터 중간 단계의 디렉터리르 모두 표시해야 함
     - 특정 위치를 가리키는 절대 경로명은 항상 동일

## 상대 경로

 - 현재 디렉터리를 기준으로 경로명을 시작
     - 매번 경로명이 동일하지 않음
     - / 이외의 문자로 시작해야 함
     - 현재 디렉터리 위치에서 서브 디렉터리로 내려갈 경우 서브 디렉터리명을 추가
     - 현재 디렉터리 위치에서 상위 디렉터리로 이동하려면 ..(마침표 2개)를 추가
     - 상대 경로명은 현재 위치한 디렉터리의 위치에 따라 달라질 수 있음

# 3. 파일과 디렉터리명 생성 규칙

    - '/'는 파일명이나 디렉터리명에 사용할 수 없음
    - 파일과 디렉터리 이름에는 알파벳, 숫자, -, _, .(마침표)만 사용할 수 있음
    - 공백, *, |, ", ', @, #, $, ^, & 등은 사용할수 없음
    - 알파벳 대/소문자는 엄격하게 구별되어 다른 글자로 취급됨
    - 파일과 디렉터리명을 선언할 때 점(.)으로 시작하면 숨김 파일로 간주됨
