# 1. 디렉터리 생성

 - 쓰기 권한을 부여 받은 디렉터리에는 새로운 디렉터리를 생성할 수 있음 
 - 쓰기 권한이 없으면 새로운 디렉터리 생성 불가
 - 새로운 디렉터리 생성 make directory의 약어인 mkdir 명령 사용

mkdir 
   
   - 기능 : 새로운 디렉터리 생성
   - 형식 : mkdir [옵션] 생성할 디렉터리명 
   - 옵션 : 
       - -p : 
             하위 디렉터리를 계층적으로 생성할 때 
             중간단계의 디렉터리가 없으면 자동으로 중간 단계 디렉터리를 생성하면서 
             전체 디렉터리를 생성

## 새로운 디렉터리 생성하기

새로운 디렉터리 생성할 때는 보통 현재의 디렉터리에 하위 디렉터리를 새로 생성하는 경우

디렉터리명은 상대 경로명이나 절대 경로명으로 지정

보편적으로 상대 경로명을 사용 디렉터리 생성

## 새로운 디렉터리 한 번에 여러 개 생성하기

## 중간 디렉터리 자동으로 생성하기 : -p 옵션

# 2. 디렉터리 삭제


 - 디렉터리  삭제 remove directory의 약어 rmdir 명령을 사용

- rmdir 
    - 기능 : 기존에 존재하는 디렉터리 삭제
    - 형식 : rmdir [옵션] 삭제할 디렉터리명
    - 옵션 
        - -p : 
           삭제하고자 하는 디렉터리를 삭제할 때 
           부모 디렉터리가 빈 디렉터리로 존재할 경우에는 부모 디렉토리도 자동으로 삭제

## 비어 있지 않은 디렉터리 삭제하기 : rm -r 디렉터리명
