이미 생성된 파일의 위치 찾기 : **find** 명령

특정 명령어 존재 위치 : **whereis**, **which** 명령

파일 관련 명령에 대한 실습을 위해 mkdir 명령으로 file_dir 디렉터리 생성

## 빈 파일 생성

빈 파일 생성 touch 명령 

 - 인자로 주어진 파일이 존재할 경우 
 - 현재 시각 또는 옵션에서 지정한 시각을 적용
 - 파일 관련 시각을 변경해 줌
 
 - touch명령의 리턴 코드는 존재 하지 않음
 - 작성되지 않은 파일을 포함해서 시각이 수정된 파일의 수를 의미

touch
    
    - 기능 : 빈 파일을 생성하는 명령
    - 형식 : touch [-acm] [r ref_file | -t time]
    - 옵션 
        -a : 최근에 작업한 파일의 접근시각만 변경 적용
        -c : 존재하지 않는 파일은 새로 만들지 않음
        -m : 최근에 작업한 파일의 수정시각만 변경 적용
        -t : [[CC]YY]MMDDhhmm[.ss] : 시각을 직접 입력

# 2. 파일 복사

파일이나 디렉터리를 복사할 때는 copy의 약어인 cp 명령을 사용


cp
 - 기능 : 디렉터리 또는 파일 복사
 - 형식 : cp [옵션] [첫 번째 인자] [두 번째 인자]
 - 옵션 
     - -i : 복사할 파일명 또는 디렉터리명이 존재하면 덮어쓸 것인지 물어보는 방법
     - -r : 디렉터리를 복사할 때 사용

## 원본 파일을 파일로 복사

cp명령 

사용 첫 번째 인자와 두 번째 인자 모두 파일인 경우
    
    - 파일과 파일을 복사하라는 의미
    
두 번째 인자로 지정한 파일명이 존재하지 않으면 원본 파일을 복사하여 생성

이미 존재하는 파일일 경우 원본 파일의 내용을 덮어쓰기를 수행

기존 파일에 존재하던 원래 내용은 모두 없어지게 됨 

### cp 명령으로 /etc 디렉터리에 존재하는 hosts 파일을 text_01 파일명으로 복사

## 여러 개의 원본 파일을 디렉터리로 복사

- 여러 개의 원본 파일을 한꺼번에 복사하기 위해서는 반드시 원본 파일을 복사될 디렉터리로 지정해 주어야 파일 복사가 가능

#### cp명령으로 etc디렉터리에 이미 존재하는 2개 (hosts, services)파일을 temp 디렉터리로 한꺼번에 복사

## 디렉터리 복사

디렉터리는 디렉터리로만 복사할 수 있음

디렉터리 복사하기 위해서는 디렉터리가 복사될 디렉터리가 존재해야 함

만약 존재 하지 않는다면 -r 옵션을 사용

새로운 디렉터리 생성과 복사가 함께 수행되도록 하면 됨

#### cp명령으로 현재 디렉터리에 존재하는 temp 디렉터리를 temp_cp디렉터리로 복사

#### cp명령과 함께 옵션 -i를 사용하게 되면 기존 파일을 덮어 씌울 것인지 아닌지
