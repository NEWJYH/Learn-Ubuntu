# 1. 파일 찾기

 
  파일을 찾고자 할 때는 find 명령 사용
  - find 명령
      - 파일 생성 일자
      - 파일명
      - 파일 소유자 등 다양한 조건에 맞는 파일을 찾을수 있음

- find
    - 기능 : 검색 조건에 맞는 파일을 지정한 위치에서 찾음
    - 형식 : find 경로 검색조건 [옵션] [동작]
    - 옵션 :
        - -name : 파일 이름으로 검색
        - -type : 파일 타입으로 검색
        - -user : 로그인 아이디로 사용자가 소유한 모든 파일 검색
        - -perm : 지정한 접근 권한과 일치하는 파일 검색
        - -exec 명령 {} \ : 검색된 파일에 명령을 실행
        - -ok 명령 {} \ : 사용자의 확인을 받아서 명령을 실행
        - -print : 검색된 파일의 절대 경로명을 화면에 출력(기본 동작)
        - -ls : 검색결과를 긴 목록형식으로 출력

### step01 
 - find 명령과 옵션 -name을 사용하여 /usr 디렉터리에 존재하는 ls 파일을 찾음

## 3. 파일의 특정 문자열 검색

- 파일 내에서 특정 문자열을 검색할 때는 grep 명령을 사용합니다.

- grep
    - 기능 : 찾고자 하는 문자열 검색
    - 형식 : grep [옵션] [찾을 문자열] [파일명]
    - 옵션 :
         - -c : 찾을 문자열이 속한 행의 개수 출력
         - -H : 파일명과 함께 출력
         - -i : 대소문자를 구별하지 않고 출력
         - -ㅣ : 해당패턴을 포함하는 파일 이름 출력
         - -n : 찾으려는 문자열이 속해 있는 행의 번호를 함께 출력
         - -r : 현재 경로부터 하위 경로까지 검색하여 출력
         - -v : 찾으려는 문자열이 없는 행 출력
         - -w : 패턴 표현식을 하나의 단어로 취급하여 검색

### step01
 - cp명령으로 /etc/services 파일을 ex_data로 복사 후 ls 명령으로 확인

### step02

   행 번호와 함께 출력

# 5. 파일 이동

 - 파일이나 디렉터리를 다른 디렉터리로 이동하기 위해서는 move 약어인 mv 명령을 사용

 - mv
    - 기능 : 파일이나 디렉터리를 이동 또는 파일명이나 디렉터리 명 변경
    - 형식 : mv [원본 파일명] [이동할 파일명]
    - 옵션 : -i : 이동할 디렉터리에 동일한 파일명이나 디렉터리가 존재하는지 메시지 출력

## 파일을 파일로 이동

 - mv명령을 사용 원본 파일을 이동하고자 하는 디렉터리에 동일한 이름의 파일이 존재할 경우
 - 원본 파일의 내용으로 덮어쓰고 기존의 내용은 삭제됨
 - 동일한 이름의 파일이 존재하지 않는다면 새로운 파일로 생성됨

### step01 
 - ls 명령으로 현재 디렉터리에 존재하는 파일 목록을 출력합니다.

### step02
 - mv 명령으로 text01파일을 ext02파일로 이동합니다.

### step03
- ls 명령으로 파일 목록을 출력하면 당초 있던 원본 파일인 text_01 파일은 없어지고
  text_02파일이 생성된 것을 확인할 수 있습니다.
