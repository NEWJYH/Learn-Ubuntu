# 1. 파일 삭제

- remove의 약어인 rm 명령을 사용
- 디렉터리를 삭제할 때 반드시 옵션 -r을 함께 사용해야 함

- rm
    - 기능 : 파일 또는 디렉터리 삭제
    - 형식 : rm [옵션] [파일명 또는 디렉터리명]
    - 옵션 
        - -i : 삭제할 파일명 또는 디렉터리명이 존재하면 삭제할 것인지를 물어보는 방법
        - -r : 지정한 디렉터리를 삭제할 때 사용 (특히 비어있지 않은 디렉터리를 삭제)
        - -ri : 지정한 디렉터리를 삭제할 때 내용을 하나씩 확인하면서 삭제
        
 - rm 명령 디렉터리 삭제할 때 물음 없이 파일을 삭제할 경우 그냥 바로 삭제되어 복구 할 수 없게 되기 때문에 신중히 삭제해야함 

 - 특히 일반 사용자가 아닌 시스템 관리자인 경우에는 더욱 신중
 - 시스템의 중요한 파일이 삭제 될 경우 전체 시스템에 심각한 영향

# 2.파일 링크

- 파일 링크 
    - 기존에 있는 파일에 새로운 파일을 연결해 주는 것으로 Windows 운영체제의 [바로가기]와
      같은 기능을 수행
    
    - 기존 파일에 새로운 파일명을 추가하여 생성하는 **하드 링크**와 
      원본을 가르키는 새로운 파일을 만드는 **심볼릭 링크**가 있음
        - **하드링크** : 기존 파일 새로운 파일명 추가 생성
        - **심볼릭 링크** : 원본을 가르키는 새로운 파일
    
    - 파일이나 디렉터리는 해당하는 하나의 I-node를 가지고 있음
    - I-node는 하나의 파일이나 디렉터리으 ㅣ모든 정보를 갖고 있는 64byte 크기로 구성된 표
    - I-node는 파일의 어드레싱
    - I-node 번호가 같은 파일은 파일명이 다르더라도 같은 파일로 취급
            

# 

## 하드 링크 만들기: ln

- 파일에 여러 개의 이름을 명명할 수 있음
- 이때 부여하는 파일명을 하드 링크라고 함
- 하드 링크는 Link의 약어인 ln명령으로 만듬

- ln
    - 기능 : 링크 파일 생성
    - 형식 : ln [옵션] 원본 파일명 링크파일명
    - 옵션 :
        - -b : 링크 파일을 생성할 때 대상 파일이 이미 존재한다면 
               백업 파일 생성 후 링크 파일 생성
        - -d : 디렉터리에 대한 하드 링크 파일 생성을 가능하게 함
               (단, root 권한으로 수행하더라도 시스템 권한 제한으로 인해 실패 가능성 높음)
        - -f : 대상 파일이 존재할 경우 대상 파일을 지우고 링크 파일을 생성
        - -i : 대상 파일이 존재할 경우 대상 파일을 지울 것인지 확인 요청
        - -s : 심볼릭 링크 파일 생성
        - -S : 백업 파일을 생성할 경우 접미사를 지정
        - -t, --target-direcory=DIRECTORY : 링크 파일을 생성할 디렉터리 지정

### step01 
    - ls -l명령으로 현재 디렉터리에 존재하는 파일 목록을 출력

### step02
- ln 명령으로 ex_data 파일에 대한 하드 링크 파일을 생성


### step03
- ls -l 명령으로 ex_data 파일에 대한 ex_data.ln 하드 링크 파일이 생성된 것을 확인

## 심볼릭 링크 만들기 : -s 옵션

 - 하드 링크 파일은 동일한 I-node 번호가 부여되는 반면,
   심볼릭 링크 파일은 전혀 다른 I-node 번호가 부여
   
 - 원본 파일과 심보릭 링크 파일은 서로 다른 파일로 취급된다는 것을 의미

### step01

- ls -i 명령으로 현재 디렉터리에 존재하는 파일의 I-node 번호를 확인

### step02
- ln -s 명령으로 현재 디렉터리에 존재하는 ex_data파일에 대한 심볼릭 링크 파일
  ex_data.sl을 생성

### step03
- ls -l 명령으로 ex_data.sl 심볼릭 링크 팡닐에 대한 상세한 내용을 출력
