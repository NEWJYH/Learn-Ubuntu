# 1. 에디터 작업 모드


 - vi 에디터의 vi는 visual의 약어
 - 사용자는 vi에디터 사용함에 있어 커서를 이동시키면서 내용을 수정
 - 작업은 명령 모드, 입력 모드, 라인 모드의 세가지 모드로 구분되어 작업 모드가 전환되면서 사용

- vi 에디터는 사용되는 명령어 대문자와 소문자 엄격 구별

- 명령모드에서 입력 모드로 전환하기 위해서는 i, I, a, A, o, O 중에서 알파벳 하나를 입력하면 됨

- 입력 모드에서 명령모드로 이동하려면 'esc'키를 눌러 이동

# 2. vi 에디터 편집 키

- 입력모드에 명령키
    - i : 커서의 위치부터 입력
    - I : 커서 행의 맨 앞부터 입력
    - a : 커서의 다음 위치부터 입력
    - A : 커서 행의 마지막부터 입력
    - o : 커서의 다음 행에 입력
    - O : 커서의 앞 행에 입력
    - s : 커서 위치의 한 글자를 지우고 입력
    - S : 커서의 행을 지우고 입력

- 명령 모드에서 커서 이동키
 - gg : 첫 행으로 이동
 - 숫자G : 숫자의 행으로 커서 이동
 - G : 마지막 행으로 이동
 - :숫자 : 숫자의 행으로 커서 이동
 - h : 커서를 왼쪽으로 한칸 이동
 - l : 커서를 오른쪽으로 한칸 이동
 - k : 커서를 위쪽으로 한칸 이동
 - j : 커서를 왼쪽으로 한 칸 이동
 - Ctrl + B : 이전 화면으로 커서를 이동
 - Ctrl + F : 다음 화면으로 커서를 이동
 - ^ : 현재 행의 처음으로 커서를 이동
 - $ : 현재 행의 마지막으로 커서를 이동ㅁ

- 명령 모드에서 문자열을 찾거나 명령 취소, 복사, 붙여넣기, 삭제 등 
  내용 편집할 때 사용하는 키

- 명령모드에서 내용편집 키

    - yy : 현재 커서의 행을 복사
    - p : 복사 내용을 현재 행 뒤에 붙여 넣음
    - x : 현재 커서 위치의 글자 삭제
    - dd : 현재 커서 위치의 행 삭제
    - r : 현재 커서 위치의 글자 수정
    - u : 수행한 명령 취소
    - ?문자열 : 커서 위치의 위쪽으로 문자열 검색
    - 숫자yy : 현재 커서부터 숫자만큼 행을 복사
    - P : 복사 내용을 현재 행 앞에 붙여 넣음 
    - X : 현재 커서 위치의 앞 글자 삭제
    - 숫자dd : 현재 커서부터 숫자만큼 행 삭제
    - cc : 커서가 위치한 행의 내용 모두 수정
    - U : 해당 행에서 수행한 명령 취소
    - /문자열 : 커서 위치의 아래쪽으로 문자열 검색

 - 특정 문자열 치환 키
     - :s/문자열1/문자열2/ : 커서가 위치한 행에서 
                            첫 번째로 나오는 문자열 1을 문자열2로 치환
     - :%s/문자열1/문자열2/g : 파일 전체에서 모든 문자열1을 문자열2로 치환
     
     - :<범위>s/문자열1/문자열2/ : 범위 내 모든 행의 각행에서 첫 번째로 검색되는 
                                  문자열1을 문자열 2로 치환
     - :<범위>s/문자열1/문자열2/g : 범위 내 모든 행에서 문자열1을 문자열2로 치환
     - :<범위>s/문자열1/문자열2/gc : 범위 내 모든 행에서 문자열1을 문자열2로 
                                    치환할 때 수정할지에 대한 여부 확인

- vi 에디터에서 저장과 종료 시 명령키
 - 라인 모드
     - q : vi 에디터에서 작업한 내용이 없으면 그냥 종료
     - q! : 작업한 내용을 저장하지 않고 종료
     - :w[파일명] : 파일명을 지정하면 새 파일로 저장 아니면 그냥 저장
     - :wq 또는 wq! : 작업한 내용을 저장하고 vi에디터를 종료
  - 명령모드
     - ZZ (Shif+Z Z) : 작업한 내용을 저장하고 vi에디터를 종료
