# 2. 파일의 속성 살펴보기

 - 1. -
 - 2.rw-rw-r--
 - 3. 1
 - 4. joung5846
 - 5. joung5846
 - 6. 54
 - 7. 1월 15 15:03
 - 8. access.txt

- 파일 속성에 대한 상세한 정보

    - 1. `-` : 파일의 종류 (-는 일반파일, d는 디렉토리)
    - 2. rw- r-- r-- : 파일을 읽고(rw-), 쓰고(r--), 실행(r--)_할 수 있는 접근 권한
    - 3. 1 : 하드 링크의 수
    - 4. joung5846 : 파일 소유자의 로그인 ID
    - 5. joung5846 : 파일 소유자의 그룹 이름
    - 6. 54 : 파일의 크기(byte 단위)
    - 7. 1월 15 15:03 : 파일을 최종 수정한 일자와 시각
    - 8. access.txt : 파일명
