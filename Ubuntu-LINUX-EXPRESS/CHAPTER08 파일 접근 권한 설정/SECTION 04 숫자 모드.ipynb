# 1. 숫자 모드 접근 권한

 - 심볼릭 모드는 파일 소유자 또는 그룹과 기타 사용자에 대한 각각의 권한 부여 제거 편리
 - 전체적으로 접근 권한 조정시 문자 조합이 다소 많아 복잡해짐

- 숫자 모드를 활용 소유자, 그룹, 기타 사용자의 권한을 한 꺼번에 설정하는 방법

- 숫자모드 접근 권한
    - 소유자 권한 : 0 ~ 700
    - 그룹 권한 : 0 ~ 070
    - 기타 사용자 권한 : 0 ~ 007

소유자 권한(u)
- 읽기, 쓰기, 실행
- 400 + 200 + 100

그룹 권한(g)
- 읽기, 쓰기, 실행
- 040 + 020 + 010

기타 사용자 권한(o)
- 읽기, 쓰기, 실행
- 004 + 002 + 001

# 2. 숫자 모드로 권한 변경

 - chmod 명령을 사용하여 숫자 모드로 ㅂ파일 접근 권한 변경
 

### step01 

- ls -l 명령으로 현재 디렉터리 존재 파일 접근 권한 확인

### step02

- chmod 666 access.txt 명령 수행 후 상세 정보 출력
