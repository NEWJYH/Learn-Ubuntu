 - 우분투 기본적 여러 사용자가 접속
 - 각 사용자를 등록하고 관리하는 방법에 대해 알아야함
 - 여러 사용자가 접속해서 우분투를 사용하려면 사용자 계정을 발급받아야 함
 
 - 관리자 계정 사용자 계정 생성과 권한 부여 등 우분투 시스템의 전체적인 관리를 수행
 
 - 관리자 계정으로 접속
     - `$`프롬프트를 `#`프롬프트로 변경
     - 관리자인 root 계정으로 로그인 해야함

# 1. 접속 계정 변경

- 우분투 시스템 접속된 사용자 계정에서 다른 사용자 계정으로 교체할 때 su 명령 사용

- su
    - 기능 : 다른 계정으로 변경
    - 형식 : su [옵션] 계정 이름
    - 옵션 
        - -c : 셸을 실행하지 않고 주어진 명령만 실행
        - -s : 지정된 셸로 접속
        - `-`, -l : 지정한 사용자의 환경변수를 적용하여 접속

# 2. 관리자 권한 명령
 - sudo 명령어는 substitute user do의 약어로 사용자 계쩡으로 접속된 상태에서
   관리자 계정인 root의 권한을 빌려서 명령을 수행할 때 사용
   
 - sudo
     - 기능 : 다른 사용자 계정의 권한으로 명령을 수행
     - 형식 : sudo [옵션] 계정 이름
     - 옵션 :
         - -i : /root 디렉터리로 이동
         - -s : 현재 디렉터리를 유지
