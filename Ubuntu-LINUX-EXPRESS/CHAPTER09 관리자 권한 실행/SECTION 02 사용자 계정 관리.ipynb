# 1. 사용자 계정 생성

- 사용자 계정 관리는 우분투 시스템 관리자의 중요한 업무
- 사용자 계정을 생성할 때는 **adduser 명령과 useradd 명령** 제공


 - adduser
     - 기능 : 사용자 계정을 새로 생성
     - 형식 : adduser [옵션] [생성할 계정 이름]
     - 옵션 :
         - -c : 사용자 이름 등 부가적인 설명 지정
         - -d : 사용자 계정의 홈 디렉터리 지정
         - -D : 기본 설정값을 출력하거나 설정
         - -e : 유효기간 설정(YYYY-MM-DD)
         - -f : 사용자 계정의 유효일자(-f-175 : 앞으로 175일간 사용 가능) 
         - -g : 사용자 계정의 로그인 기본그룹 GID 지정
         - -G : 2차 그룹의 GID 지정
         - -o : 사용자 계정의 UID 중복 허용
         - -p : 사용자 계정의 패스워드
         - -s : 사용자 계정의 기본 셸 지정
         - -u : 사용자 계정의 UID 지정
         


- 일반 사용자 계정 생성은 오직 관리자만이 발급

#### etc/passwd 파일에 저장되어 있는 정보 살펴보기
- 1. member_001
    - 로그인 ID
        - 사용자 계정의 이름을 의미

- 2. x
    - 암호 저장 항목
        - 이전 프로그램과의 호환성을 위해 기본항목을 그대로 유지하고 있음 x 로 표시
- 3. 1001
    - UID
        - 우분투 시스템이 사용자를 구분하기 위해 사용되는 사용자 ID 번호
        - 일반사용자 1000부터 할당
        - 사용자에 대한 시스템 관리업무를 위해 내부적으로 사용자를 위해 예약
        - 임의로 변경하지 않는 것이 좋음
        - 사용자를 구별할 때 로그인 ID는 서로 다르더라도 UID가 같으면 
          우분투 시스템은 동일한 사용자로 판단
        - 시스템 관리자 UID가 동일하지 않게 관리해야함
- 4. 1001
    - GID
        - 사용자가 속해 있는 그룹 ID
        - 우분투에서 사용되는 모든 사용자는 무조건 하나 이상의 그룹에 속해 있어야 함
        - GID는 사용자를 등록할 때 정해짐
        - GID를 등록하지 않으면 자동으로 로그인 ID가 그룹으로 등록
        - 등록된 GID는 /etc/group 파일에서 확인할 수 있음
- 5. ,,,
    - 설명
        - 사용자에 대한 일반적인 정보(이름, 부서, 연락처 등)가 기록되는 항목
        - 유닉스에서 유래된 기원을 바탕으로 사용자 개인정보라고도 불림
- 6. home/member_001
    - 홈 디렉터리
        - 사용자 계정에 할당된 홈 디렉터리는 절대 경로의 형태로 지정
        - 홈 디렉터리는 사용자 계정을 생성하면 자동으로 부여
        - 홈 디렉터리 아래에는 서브디렉터리와 파일드을 생성하거나 저장할 수 있음
- 7. /bin/bash
    - 로그인 셸
        - 사용자 로그인 기본 셸을 지정
        - 우분투 bash 셸을 기본으로 적용
        - 사용자 로그인 셸을 지정

# 3. 사용자 계정 정보 수정

 - 이미 생성된 사용자 계정에 대한 정보를 수정할 때는 **usermod** 명령 사용

- 로그인 ID, UID, GID, 설명, 홈 디렉터리, 기본 셸 등 계정과 관련된 모든 정보를 수정 가능

 - usermod
     - 기능 : 사용자 계정정보 수정
     - 형식 : usermod [옵션] [사용자 계정 이름]
     - 옵션 : 
         - -c : 사용자 이름 등 부가적인 설명 수정
         - -d : 사용자 계정의 홈 디렉터리 수정
         - -e : 사용자 계정 만료일(유효기간) 수정(YYYY--MM-DD)
         - -f : 사용자 계정 비활성화 일자 수정
         - -g : 사용자 계정의 로그인 기본그룹 GID 수정
         - -G : 2차 그룹의 GID 수정
         - -l : 사용자 계정 이름 변경
         - -o : 사용자 계정의 UID 중복 허용
         - -s : 사용자 계정의 기본 셸 수정
         - -u : 사용자 계정의 UID 수정

# 4. 사용자 계정 삭제

- 관리자 계정에서 생성한 사용자 계정을 삭제할 때는 **userdel** 명령을 사용

- 관련된 홈디렉터리와 파일들을 모두 삭제하는 것이 바람직
- 필요한 경우 계정만 삭제할 때도 있지만 가급적 디스크에 남겨진 파일이 존재하기 때문에 
  한꺼번에 삭제하는 것이 좋음
  
  
- userdel
    - 기능 : 사용자 계정 삭제
    - 형식 : userdel [옵션] 계정 이름
    - 옵션 :
        - -r : 홈 디렉터리 삭제
        - -f : 사용 중인 계정 이름도 강제 삭제
        - -rf : 홈 디렉토리 이외에 존재하는 파일 삭제

## 사용자 계정 삭제 수행 방법

 - 사용자 계정의 홈 디렉터리를 root 또는 다른 소속의 그룹으로 속성을 변경하였을 경우
   홈 디렉터리를 삭제하기 위해서는 **groupdel 명령**을 사용하여 
   `강제로 사용자 계정의 홈 디렉터리를 삭제`해야 함

 - 사용자 계정을 생성하게 되면 사용자 계정에 대한 각종 정보는 /passwd, /etc/group,
   /etc/shadow, /var/spool/mail 파일에 자동으로 생성

 - 만일 삭제한 동일 계정 이름으로 다시 사용자 계정을 생성한다고 할 때
     - 기존에 생성했던 사용자 계정에 대한 모든 정보를 깨끗하게 삭제하지 않은 상태에서
     - 같은 이름의 사용자 계정이름을 사용하려고 하면 오류 메시지 출력
     - 사용자 계정을 삭제할 때는 사용자 계정만 삭제해서는 안됨

 - 사용자 계정 삭제하는 과정
     - 1. /etc/passwd 파일에서 사용자 계정 삭제 : userdel -r 계정 이름
     - 2. /etc/group 파일에서 사용자 그룹 삭제 : groupdel 계정 이름
     - 3. /home 디렉터리에 있는 사용자의 홈 디렉터리 삭제 : rm -rf /home/계정 이름
     - 4. /var/spool/mail에 있는 사용자 메일 삭제 : rm /var/spool/mail/계정 이름
     - 5. 데이터베이스 연동에서 삭제하려는 계정 이름과 암호 삭제
     - 6. 기타 응용 프로그램에 존재하는 사용자 계정의 정보 삭제
     - 7. 우분투 시스템을 재시작

## UID를 이용하여 사용자 계정 관련 모든 파일 삭제 방법

- 사용자 계정을 생성할 떄 자동으로 생성되었던 여러 파일을 일일이 찾아서 삭제하는 과정 다소 복잡

- 한 번에 사용자 계정을 삭제하려면 파일 찾기 명령인 find 명령을 사용하면 됨

- find 명령으로 UID를 찾아서 사용자 계정과 관련된 모든 파일을 한꺼번에 삭제

- find -user UID -exec rm -r {} \;
    - 기능 : 파일내용 중에서 UID를 찾아서 관련된 파일을 모두 삭제
    - 형식 : find [옵션] UID [옵션] [명령] {} \;

# 5. 기타 명령어

- su : 다른 계정으로 변경
- useradd : 사용자 계정 생성
- passwd : 사용자 계정의 암호 수정
- tail : 파일 내용을 마지막 라인부터 10번쨰 라인까지만 내용 출력 
- usermod : 사용자 계정정보 수정
- grep : 찾고자 하는 문자열 검색
- sudo change : 시스템 보안을 위해 사용자 계정의 암호 기한 설정 및 변경
- sudo userdle : 사용자 계정 삭제
- find : 파일 내용 찾기
- groupadd : 새 그룹 생성
- gpasswd : 그룹 암호 가ㅘㄴ리
- sudo cat : 우분투 시스템에 등록되어 있는 정보 출력
- id : 현재 사용자의 실제 id와 유효 사용자 ID, 그룹 ID를 출력
- newgrp : 소속 그룹 변경
- sudo chown : 파일이나 디렉터리의 소유권 및 그룹 소유권을 변경
- sudo chgrp : 파일이나 디렉터리의 그룹 소유권을 변경
- shutdown : 우분투 시스템 종료
