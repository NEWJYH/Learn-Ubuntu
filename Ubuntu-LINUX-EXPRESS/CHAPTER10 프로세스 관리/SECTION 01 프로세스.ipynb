# 1. 프로세스의 개념

 - 하드 디스크에 저장된 **프로그램**을 
   **CPU의 명령**에 따라 **메모리에 로딩**되어 **활성화** 된 것을 **프로세스**라고 함

# 2. 프로세스 상태 전이

 - 프로세스는 운영체제의 스케줄링에 의해 상태가 전이

 - 하나의 프로그램이 실행되면 
   그 프로그램에 대응되는 프로세스가 생성되어 준비 리스트의 끝에 들어감
   
 - 준비 리스트 상의 다른 프로세스들이 CPU를 할당받아 준비 리스트를 벗어나면 
   그 프로세스는 점차 준비 리스트의 앞으로 나가게 되고 할당된 순서가 도래되면 CPU를 
   사용할수 있게 됨

# 3. 프로세서 번호와 작업 번호


## 프로세스 번호

- CPU에서 처리하고자 하는 작업이 
  멀티태스킹과 같이 여러 개의 프로세스가 메모리에 로딩되어 있을 경우 
  프로세스를 구분하려면 각각의 프로세스에 대한 고유 식별방법이 존재
  
- CPU가 프로세스를 구분하기 위해 부여되는 고유번호를 바로 프로세스 번호라고 하며
  각 프로세스가 가지는 고유의 번호를 PID라고 함
 

## 작업 번호

 - 현재 실행되는 백그라운드 프로세스가 CPU를 점유하여 
   작업을 수행할 때의 순차 번호를 작업 번호라고 합니다.

# 4. 프로세스 서비스와 관계

## 데몬 프로세스

 - 우분투에서 데몬 프로세스를 통해 다양한 서비스를 제공
 - 데몬 프로세스는 우분투에서 특정 서비스를 제공하기 위해 존재하는 프로세스
 - 평상시에는 대기 상태로 있다가 서비스 요청이 들어오면 해당하는 서비스를 제공

## 부모-자식 관계의 프로세스

- 부모 프로세스는 PPID를 가지고 자식 프로세스는 PID를 가지고 있음
- 부모 프로세스인지 자식 프로세스인지 구분이 됨
- 부모 프로세스를 종료하게 되면 종속된 자식 프로세스 또한 강제로 종료

# 5. 고아 프로세스와 좀비 프로세스

## 고아 프로세스

 - 고아 프로세스는 부모-자식 관계의 프로세스에서 자식 프로세스가 종료되지 않은 상태에서
   부모 프로세스가 종료되는 경우를 의미
   
 - 이렇게 발생한 고아 프로세스의 종료는 1번 프로세스를 부모 프로세스로 의존하게 되어 
   작업이 무난히 마치고 프로세스를 종료하게 됨

## 좀비 프로세스

 - 좀비 프로세스 발생되는 이유는 부모 프로세스가 자식 프로세스의 종료를 기다린 뒤에
   종료를 처리해 주는 정상적인 과정을 거치지 않고 부모 프로세스를 종료할 경우 발생
   
 - 좀비 프로세스는 kill 명령으로 제거할 수 없음

#### 좀비 프로세스를 제거하기 위해서는 SIGCHLD 시그널을 부모 프로세스에 보내 부모 프로세스가 자식 프로세스를 정리하도록 수행하거나 부모 프로세스 자체를 종료해야만 좀비 프로세스를 제거할 수 있음
