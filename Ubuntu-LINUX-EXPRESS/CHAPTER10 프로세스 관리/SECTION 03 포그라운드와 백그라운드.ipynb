# 1. 포그라운드 작업

- 터미널 창에서 명령어를 입력하면 셸은 사용자가 입력한 명령을 수행
- 그 결과를 화면에 보여줌

- 사용자는 화면에 나타난 출력결과를 보고 다른 명령을 입력하는 대화식으로 수행하는 작업

- 즉 사용자와 상호작용을 하여 작업을 수행하도록 해 주는 프로세스를 포그라운드 작업이라고 함

# 2. 백그라운드 작업

- 백그라운드 작업은 프로세스는 실행되었지만 직접 눈으로 프로세스가 확인되지 않는 작업

- 백신 프로그램, 서버 데몬 등과 같이 화면에 나타나 눈에 보이지는 않지만
  실행되는 명령을 백그라운드 작업이라고 하고 
  이때 수행하는 프로세스를 백그라운드 프로세스라고 합니다.
  


### step01

 - 프로세스가 잠시 쉬도록 sleep 명령으로 포그라운드 작업을 수행하게 되면
   작업이 끝날 때까지 기다려야 하므로 다른 작업을 수행할 수 없게 됨
   


### step02

 - 터미널 창에서 sleep 명령과 &(앰퍼샌드) 기호를 선언하여 백그라운드 작업을 30초 동안 수행
 - 백그라운드 작업 중에는 다른 작업을 수행할 수 있음

---

# 3. 작업 제어

 - 포그라운드 작업을 백그라운드 작업으로 전환하거나 
   백그라운드 작업을 포그라운드 작업으로 전환하는 제어를 의미
   
## 작업 목록 : jobs

 - 현재 실행 중인 백그라운드 작업의 목록을 출력할 때는 jobs 명령을 사용

 - jobs
     - 기능 : 백그라운드 작업 목록 출력
     - 형식 : jobs [%작업 번호]
     %작업번호 
         - %번호 : 해당 작업의 정보만 출력
         - %+ 또는 %% : 작업순서가 +인 작업 정보를 출력
         - %- : 작업순서가 -인 작업 정보를 출력

### step01
 - 백그라운드 작업과 포그라운드 작업을 수행

### jobs 명령의 출력 항목

- 항목
    - 작업 번호 
        - [2] 
            - 백그라운드 작업 번호를 나타냄
    - 작업 순서
        - + : 
            - +: 가장 최근에 접근한 직업
            - :+ 작업 전에 접근한 작업
            - 공백 : 그 이외의 작업
        
    - 작업 상태
        - 실행중 :
            - 실행중 : 현재 실행(Running)
            - 완료됨 : 작업 정상 종료(Done)
            - 종료됨 : 비정상 작업 종료(Terminated)
            - 정지됨 : 작업 잠시 중단(Stopped)
    - 명령
        - sleep 30& :
            - 백그라운드로 실행 중인 명령

## 작업 전환
    - 작업 전환이란 현재 포그라운드로 실행 중인 작업을 백그라운드 작업으로 전환하거나 
      백그라운드로 실행 중인 작업을 포그라운드 작업으로 전환하는 것을 의미
      
      


- 작업 전환 명령
 - 명령 : 
     컨트롤 z 또는 stop [%작업번호] : 포그라운드 작업을 잠시 중단
     bg[%작업 번호] : 작업 번호가 지시하는 작업을 백그라운드로 전환
     fg[%작업 번호] : 작업 번호가 지시하는 작업을 포그라운드로 전환

### step01 

- 터미널 창에서 jobs 명령으로 작업 목록을 확인

### step02

- sleep 명령으로 포그라운드 작업을 수행한 다음 Ctrl + Z를 눌러 작업을 일시 중단

### step03

- 현재 중단된 포그라운드 작업을 bg[%작업 번호] 명령으로 백그라운드 작업으로 전환
- 작업 번호는 지정하지 않고 bg명령만 사용할 경우 작업순서가 +인 작업에 대해서만 적용

### step04

 - jobs 명령으로 현재 수행 중인 백그라운드 작업 목록을 확인


### step05

- fg[%작업 번호] 명령으로 백그라운드 작업을 포그라운드 작업으로 전환한 다음 jobs 명령으로
  현재 작업 중인 목록을 출력
  
- bg 명령과 마찬가지로 작업 번호는 지정하지 않고 fg명령만 사용할 경우 작업순서가 + 인 작업
  에 대해서만 적용

### 포그라운드 작업 종료 : Ctrl + C

## 백그라운드 작업 종료 : kill

 - 백그라운드 작업은 kill 명령을 사용하여 강제로 종료
 - kill 명령을 사용할 때 인자로 PID 대신 인자로 [%작업 번호]를 선언 해도 됨
 - 현재 수행 중인 백그라운드 작업을 강제로 종료
