 - 우분투는 소프트 웨어를 소스코드 형식 또는 바로 설치하여 사용할수 있는 패키지 형태 제공

# 1. APT 명령

- APT(Advanced Package Tool) 
    - 레드햇 계열 리눅스 Fedora와 CentOS 리눅스의 Yum명령과 동일한 명령 수행
    - APT의 대표적인 명령 apt-cahe와 apt-get 명령을 사용
    
    - Ubuntu 18.04 LTS 버전부터 apt-cache와 apt-get 일부를 통합한 apt 명령도 선보임
    - apt-cache와 apt-get 명령을 apt 명령으로 모두 대체하여 사용할수 있는 상황은 아님
    - apt-cashe와 apt-get 명령을 모두 사용하도록 함 

## apt-cashe 명령

- apt-cache 명령은 패키지 데이터베이스(APT 캐시)에서 패키지와 관련된 정보를 출력

 - apt-cache
     - 기능 : 패키지 검색 및 정보보기 명령
     - 형식 : apt-cache [옵션] [서브 명령]
     - 옵션 : 
         - -f : 패키지에 대한 전체 정보 출력
         - -h : 관련 도움말 출력
     - 서브 : 
         - stats : 캐시 통계 정보(전체 패키지 이름과 단일 가상 패키지 등) 출력
         - dump : 현재 설치되어 있는 패키지 업그레이드
         - pkgnames : 사용 가능한 모든 패키지의 이름 출력
         - search [키워드] : 키워드에 해당하는 패키지 검색
         - show [패키지 이름] : 해당 패키지의 정보를 출력
         - showpkg [패키지 이름] : 패키지의 의존성과 역의존성에 대한 정보 출력

## APT 캐시 통계정보 출력 : stats

 - apt-cache stats 명령 
     - 시스템에 설치되어 있는 전체 패키지의 이름과 일반 패키지,
       순수 가상과 단일 가상 및 혼합 가상 패키지 등의 APT 캐시의 전반적인 통계정보를
       출력하는 명령
       
     - apt-cashe stats 명령으로 우분투에 설치되어 있는 APT 캐시의 통계정보를 확인

## 패키지 업데이트 : dump

- apt-cache dump 명령으로 현재 시스템에 설치되어 있는 모든 패키지에 대한 업데이트를 수행할 수 있음

```bashshell
// bashshell
apt-cache dump

```

## 모든 패키지 이름 출력: pkgnames

## 패키지 정보 출력 : show

## 패키지 설치 및 업데이트 명령 : apt-get

- apt-get 명령은 우분투를 포함한 데비안(Debian) 계열의 리눅스에서 패키지 설치 및 업데이트를 수행하는 명령

 - 파티션 패키지인 GParted 프로그램을 설치하여 위해 
 - apt-get(Advanced Packaging Tool) 명령어 사용

 - apt-get
     - 기능 : 패키지 설치 및 업데이트 명령
     - 형식 : apt-get [옵션] [서브 명령]
     - 옵션 :
         - -d : 패키지 다운로드 수행
         - -f : 의존성이 깨진 패키지 수정
         - -h : 관련 도움말 출력
      - 서브 :
         - update : 패키지 저장소에서 새로운 패키지 정보를 가져옴
         - upgrade : 모든 패키지를 최신 버전으로 업그레이드
         - install [패키지 이름] : 해당 패키지 설치
         - remove [패키지 이름] : 설치된 해당 패키지 제거
         - purge [패키지 이름] : 설치된 해당 패키지와 설정 파일을 모두 삭제
         - autoremove : 시스템에 설치된 패키지를 자동으로 정리 및 삭제
         - download [패키지 이름] : 해당 패키지를 현재 디렉터리에 다운로드
         - autoclean : 오래된 패키지 또는 불완전한 다운로드 패키지 제거
         - check : 의존성이 깨진 패키지 확인
         - clean : /var/cache/apt/archives 디렉터리에 캐시되어 있는 모든 패키지를 제거하여 디스크 공간 확보

## 패키지 정보 업데이트 : update

 - update 서브 명령을 수행하면 /etc/apt/sources.list에 새로운 패키지 정보를 가져와서
   APT 캐시를 수정
   
 - 만약 /etc/apt/sources.list 파일을 수정하였다면 반드시 이 명령을 수행해야 함
 - apt-get update 명령으로 현재 시스템에 설치되어 있는 모든 패키지에 대한 정보를 업데이트 할 수 있음

## 패키지 업그레이드 : upgrade

 - upgrade 서브 명령은 현재 시스템에 설치되어 있는 패키지 중에서 업그레이드가 필요한 
   패키지를 찾은 다음 새로운 버전이 있는 패키지에 대해서는 모두 업그레이드를 수행

``` bash
// bashshell
sudo apt-get upgrade
```

 - 우분투에 설치된 패키지 중에서 특정 패키지에 대한 업그레이드가 필요한지 알아보려면
   upgrade 서브 명령과 grep 명령으로 검색
 - 검색과정을 멈추려면 Ctrl + C 를 눌러 강제로 중지시키면 됨

## 특정 패키지 설치 또는 업그레이드 : install

- install 서브 명령은 하나 이상의 패키지를 설치하거나 업그레이드를 수행
- 패키지를 설치하는 과정 중에 설치할 것인지에 대한 물음에 y를 입력해야 패키지가 설치됨
- 하나의 패키지만 설치할 경우 다음과 같이 명령을 수행

```bashshell
// bash

sudo apt-get install [패키지 이름]
```

 - 패키지 설치할 때 y를 입력하지 않아도 자동으로 입력되도록 수행하려면 -y 옵션 입력

```shell

sudo apt-get -y install [패키지 이름]

```

 - 하나의 패키지가 아닌 여러 패키지를 한번에 설치
 
 ```shell
 
 sudo apt-get install [패키지 이름1] [패키지 이름2]
 ```

 - 패키지 설치할때 업그레이드를 수행하지 않으려면
 
 ```shell
 
 sudo apt-get install [패키지 이름] --no-upgrade
 ```

- 패키지를 설치하지 않고 업그레이드만 수행하려면 --only-upgrade 옵션을 사용

```shell
sudo apt-get install [패키지 이름] --only-upgrade
```

 ## 특정 패키지 삭제 : purge 또는 remove
 
  - remove 서브 명령만 사용하여 패키지를 삭제할 경우 해당 패키지의 설정 파일은 남아 있게 됨

- 남은 파일 까지 모두 삭제하려면 purge 서브 명령 사용

- 패키지를 삭제할 때 설정 파일을 남겨놓는 이유는 나중에 동일한 패키지를 다시 설치할 때
  재사용하기 위함
 
- 특정 패키지에 대해 설정 파일을 포함하여 패키지를 삭제할 때는 purge 서브 명령을 사용

```shell
sudo apt-get purge [패키지 이름]
```

 - remove 서브 명령과 --purge 옵션을 함께 사용할 수도 있음
 
 ```shell
     sudo apt-get remove [패키지 이름] --purge
 ```

##  패키지 자동 정리 및 삭제 : autoremove

 - 패키지를 설치하고 삭제하는 과정에서 더 이상 존재할 필요가 없는 패키지가 남아 있을수 있음 
 - 잔존해 있는 패키지에 대해서는 autoremove 서브 명령으로 자동 정리 및 삭제 수행
 
 ```shell
  sudo apt-get autoremove
 ```

## 디스크 공간 정리 : clean

 - clean 서브 명령으로 패키지를 검색하였거나 다운로드한 파일들을 삭제한 후 디스크 공간을 정리 할 수 있음

## 특정 패키지 다운로드 : download

 - 특정 패키지를 직접 설치하지 않고 다운로드만 수행하려면 download 서브 명령을 사용
 - sudo apt-get downloac 명령으로 docker.io 패키지를 다운로드한 다음 해당 파일이
   존재하면 위치를 확인하기 위해서는 다음 명령을 수행

```shell
sudo apt-get download docker.io
```

## 특정 패키지의 소스 관련 서브 명령 : source

 - source 서브 명령은 특정 패키지의 소스 코드가 필요한 경우 다운로드할 떄 사용
 - 필요한 특정 패키지의 소스 코드에 대해 다운로드를 수행하려면 다음과 같이 명령을 수행
 
 ```shell
 
     sudo apt-get --download-only source [패키지 이름]
 
 ```

 - 다운로드를 수행한 특정 패키지의 소스 코드에 대한 압축을 해제
 
 ```shell
 sudo apt-get source [패키지 이름]
 ```

 - 특정 패키지의 소스 코드를 다운로드를 수행하여 압축을 해제하고 컴파일
 ```shell
     sudo apt-get --compile source [패키지 이름]
 ```

# 2. DPKG 명령

- DPKG 명령은 래드햇 계열의 리눅스인 Fedora와 CentOS 리눅스의 RPM 명령과 동일한 명령을 수행

- DPKG 명령은 데비안 계열의 리눅스인 우분투에서 패키지를 관리하기 위해 사용

- APT 명령은 소프트 웨어 관리하기 위해 내부적으로 DPKG 명령을 이용

- 보통 패키지를 설치할 때 APT 명령만으로 충분하지만 시스템에 있는 특정 파일이 어떤 
  패키지에 포함되는지 등의 확인 작업을 수행하려면 DPKG 명령이 필요하게 됨
  
- APT명령과 차이점이 있다면 DPKG 명령은 시스템의 특정 파일이 어느 패키지에 포함되어
  있는지 등에 대한 구체적인 기능을 확인할 떄 옵션과 함께 사용
  

- dpkg
    - 기능 : 패키지 관리 명령
    - 형식 : dpkg [옵션] [패키지 이름 또는 파일 이름]
    - 옵션 :
        - -ㅣ: 설치된 패키지 목록 전체 출력
        - -ㅣ [패키지 이름] : 해당 패키지의 설치상태 정보 출력
        - -r [패키지 이름] : sudo 명령을 사용하여 해당 패키지를 삭제
        - -p [패키지 이름] : sudo 명령을 사용하여 해당 패키지와 설정 정보 모두 삭제
        - -s [패키지 이름] : 해당 패키지의 자세한 정보 출력
        - -L [패키지 이름] : 해당 패키지가 설치된 파일목록 출력
        - -S [경로명] : 경로명이 포함된 패키지 검색
        - -c [.파일 확장자] : 해당 파일 확장자를 가진 파일의 내용 출력
        - -i [.파일 확장자] : sudo 명령으로 해당 파일 확장자를 가진 파일 설치
        - -x [.파일 확장자] [디렉터리명] : 해당 파일 확장자를 가진 파일을 지정한 디렉터리에 풀어 놓음

## 패키지 목록 출력 : -l

## 패키지 검색 : search

- aptitude 명령과 update 서브 명령으로 gnome으로 시작하는 패키지를 검색하려면 다음 명령 수행

```shell
aptitude search gnome
```

## 패키지 상세정보 : show

- aptitude 명령과 show 서브 명령으로 gnome-clocks 패키지의 상세한 정보를 확인하려면 
  다음 명령을 수행

```shell
aptitude show gnome-clock
```

## 패키지 설치 : install

- aptitude 명령과 install 서브 명령으로 gnome-clocks 패키지를 설치하려면
  다음 명령을 수행
  
  ```shell
      sudo aptitude install gnome-clocks
  ```
