# 1. 파일 아카이브

 - 아카이브(Archive)란
     - 특정 시간에 함께 묶인 파일과 디렉터리의 집합을 의미
     - 다른 표현으로 저장소 또는 보관소라고도 함
     - 아카이브는 다른 시스템과 파일이나 디렉터리를 주고 받거나 파일을 백업하기 위한 용도 사용
     - 파일 아카이브를 위해 사용되는 명령은 tar, cpio, ar 등이 있음

- tar
    - 기능 : 파일과 디렉터리를 묶어 하나의 아카이브로 생성하는 명령
    - 형식 : tar 기능[옵션] [아카이브 생성 파일명] 대상 파일명
    - 기능 : 
        - c : 새로운 tar 아카이브 생성
        - r : 새로운 파일 추가
        - t : tar 파일에 존재하는 파일 리스트 출력
        - u : 수정된 파일에 대한 업데이트 수행
        - x : tar 파일에 있는 원본 파일을 추출
    - 옵션 : 
        - C [디렉터리명] : 파일들을 특정 디렉터리에 풀어 놓음
        - f [디바이스명 또는 파일명] : 아카이브 파일 또는 테이프 장치를 지정
          (디바이스명을 -으로 지정하면 tar 파일 대신 표준입력에서 읽어 들임)
        - h : 심볼릭 링크의 원본 파일 포함
        - j : bzip2로 압축 또는 해제
        - p : 파일 복구 시 원래의 접근 권한을 유지
        - v : 처리하고 있는 파일의 정보 출력
        - z : gzip으로 압축 또는 해제
        - Z : compress를 통해 아카이브를 필터링

   아카이브를 업데이트 하기 위해 사용하는 기능

## 아카이브와 파일 압축

 - 아카이브를 생성하면서 파일압축을 동시에 수행 가능
 - 디스크 공간이 부족할 때 유용하게 사용
 - tar 명령과 cvzf 옵션을 함께 사용함
 - arch_test 디렉터리를 아카이브 생성과 동시에 arch_test.tar.gz 압축 파일로 생성
 
 ```shell
 tar cvzf arch_test.tar.gz arch_test
 ```

- 파일 확장자는 .gz는 gzip형식으로 압축한 파일임을 쉽게 알아보기 위해 선언한 것
- 아카이브 생성과 동싱 bzip2 형식으로 파일을 압축하기 위해 
  tar 명령과 함께 j 옵션 사용
- arch_test 디렉터리를 arch_tst.tar.bz2 파일명으로 생성하려면 다음과 같이 명령 수행

```shell

tar vcjf arch_test.tar.bz2 arch_test
```

# 2. 파일 압축과 해제

 - 압축: 파일의 용량을 줄여주는 기능을 의미
 - 해제 : 압축된 파일을 해제하여 원래의 파일 용량으로 되돌려 주는 기능

# 

## gzip 명령으로 파일 압축

 - gzip
     - 기능 : gip 형식으로 파일 압축 명령
     - 형식 : gzip [옵션] 파일 이름
     - 옵션 
         - -d : 파일 압축 해제
         - -l : 압축된 파일의 정보 출력
         - -r : 하위 디렉터리 이동하여 파일 압축
         - -t : 압축 파일 검사
         - -v : 압축 정보를 화면에 출력
         - -9 : 최대한 압축

## 압축 파일 풀기 : gunzip

## bzip2 명령으로 압축

- bzip2 명령은 gzip 명령에 비해 압축률은 좋지만 속도가 느리다는 단점이 있음

 - bzip2
     - 기능 : bzip2 형식으로 파일 압축 명령
     - 형식 : bzip [옵션] 파일 이름
     - 옵션 :
         - -d : 파일 압축 해제
         - -l : 압축된 파일의 정보 출력
         - -t : 압축 파일 검사
         - -v : 압축 정보를 화면에 출력
         - -best : 최대한 압축

 - bzip2 명령으로 arch_test 디렉터리를 arch_test.tar 파일명으로 압축하려면

## 압축파일 내용 보기 : bzcat
