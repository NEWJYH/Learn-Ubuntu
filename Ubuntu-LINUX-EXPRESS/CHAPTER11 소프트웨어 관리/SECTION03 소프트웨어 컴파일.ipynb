# 1. 컴파일러의 개념

 - 컴파일러(Compiler)란 자연어로 구성된 고급 언어(C, Python, Java 등)를 
   컴퓨터가 이해할 수 있는 기계어(0,1)로 번역해 주는 소프트웨어를 의미
   
 - 자연어를 기계어로 번역해 주는 과정을 컴파일(Compile)이라고 함
 
 - C 언어로 작성한 프로그램을 컴퓨터가 실행할 수 있도록 컴파일을 수행하려면
   C 컴파일러가 필요

# 2. gcc 컴파일러

 - 우분투에서 사용하는 c 컴파일러는 gcc
 - 만약 gcc가 설치되어 있지 않다면 설치

## gcc 컴파일러 설치

- 현재 사용중인 우분투에 설치되어 있는 c 컴파일러가 존재하는지 확인하기

### step01
- aptitude 명령으로 gcc 패키지가 설치되어 있는지 확인


### step02

- 만약 gcc 패키지가 설치되어 있지 않다면 apt-get 명령으로 gcc 패키지 설치

```shell
    
    sudo apt-get install gcc

```
