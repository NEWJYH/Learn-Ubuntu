 - 여러 사용자가 우분투에 접속해서 사용
 - 전체 디스크 용량이 부족하거나 사용 중인 디스크에 배드 섹터가 발생할수 있음
 - 주기적으로 디스크를 점검하는 것이 좋음
 - 우분투 디스크 사용량 확인 명령 **df명령, du명령**

# 1. 전체 디스크 사용량

 - 현재 사용 중인 디스크의 전체 사용량을 출력할 때는 df 명령을 사용
 - df명령 disk freespace의 약어
 - 현재 사용중인 디스크의 용량을 파티션별로 확인
 
 - df 명령
     - 현재 시스템의 디스크 사용량에 대해 파일 시스템의 장치명
     - 파일 시스템의 전체 용량 및 사용량
     - 파일 시스템의 추가 사용이 가능한 디스크 남은 용량
     - 사용량에 대한 퍼센트 표시
     - 마운트 포인트
     
     - 옵션없이 사용시 블록 단위는 기본적으로 1024byte
     

- df 
    - 기능 : 하드 디스크의 남은 용랴에 대한 정보를 출력
    - 형식 : df [옵션] [파일명]
    - 옵션 :
        - -a : 모든 파일 시스템에서 사용 중인 디스크의 용량에 대한 정보 출력
        - -B : 지정한 크기를 블록 단위로 디스크의 용량을 출력
        - -h : 디스크 사용량을 보기 쉬운 단위(GB, MB, KB)로 출력
        - -H : 디스크의 용량을 1024Byte가 아닌 1000Byte 단위로 출력
        - -i : inode의 남은 공간과 사용 공간, 사용 퍼센트 출력
        - -k : 디스크 사용량을 KB 단위로 출력
        - -m : 디스크 사용량을 MB 단위로 출력
        - -t 파일타입 : 지정한 파일타입의 종류에 해당하는 디스크 사용량 출력
        - -T : 파일 시스템의 형태를 추가하여 각각의 파티션 정보 출력
        - -x 파일타입 : 지정한 파일타입을 제외한 나머지 모든 파일 시스템의 정보출력

### step1 

- df 명령으로 혅 ㅐ사용 중인 우분투의 디스크 사용량에 대한 정보를 출력

### step02

- df 명령과 함께 옵션 -h를 선언하여 디스크의 블록 단위보다는 
  사람이 쉽게 이해할 수 있도록 GB와 MB 단위로 디스크 사용량을 출력

### step03

- df 명령과 함께 옵션 -T를 선언하여 디스크 사용량을 파일 타입별로 출력

### step04

- df 명령과 함께 옵션 -T 와 -h옵션을 한꺼번에 선언하여 
  파일 시스템의 타입별로 디스크 사용량에 대한 정보를 GB MB단위로 출력

# 2. 사용자별 디스크 사용량

- 디렉터리 또는 사용자별 디스크 사용량을 출력할 때는 **du 명령**

 - du (disk usage)
 - 특정 디렉터리 또는 파일 단위별 디스크 용량 확인
 - du 명령 옵션 함께 선언하지 않으면 현재 경로의 모든 디렉터리의 크기가 KB단위로 출력

- du
    - 기능 : 디렉터리 또는 사용자별 디스크 사용량 정보 출력
    - 형식 : du [옵션] [디렉터리명 또는 계정이름]
    - 옵션 : 
        - -s : 모든 파일 시스템에서 사용 중인 디스크의 용랴엥 대한 정보출력
        - -h : 디스크 사용량을 보기 쉬운 단위 KB MB KB로 출력

### step01

- pwd 명령으로 현재 위치한 디렉터리의 위치를 확인

### step02

- du 명령으로 디렉터리 또는 사용자별 디스크 사용량을 출력

### step03

- du 명령과 함께 옵션 -s를 선언하여 현재 위치한 디렉터리의 전체 디스크 사용량을 출력

### step04

- du 명령과 함께 옵션 -sh를 선언하여 디스크 사용량을 출력합니다.
