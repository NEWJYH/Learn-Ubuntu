# 1. 셸의 기능

- 셸은 사용자와 커널 사이에서 중계자 역할을 수행하는 프로그램
- 최초의 셸인 '본 셸'부터 리눅스 셸이라는 '배시 셸'까지 다양한 셸이 제공 개발 됨

- 셸은 명령어 해석기, 프로그래밍, 사용자 환경 설정 등의 기능을 사용할 수 있도록 제공해 줌

 - 하드웨어(CPU, RAM, HDD) ----------------------------
 - -----리눅스 커널(번역된 언어로 명령 수행)--------------------
 - -----------셸(명령어 해석기)-------------------------------------------
 - ---------- 응용프로그램(엑셀, 파워포인트 등) -------------------------

# 2. 셸의 종류

- 셸은 유닉스 운영체제에서 제공하기 시작

- 대부분은 유닉스 시스템에서 지원하는 주요 셸은 본 셸, C 셸, 콘셸, 배시 셸 등이 있음

- 각 셸은 일반적으로 명령을 입력하고 실행하는 기능은 비슷하지만 에일리어스 설정이나 
  초기화 파일, 스크립트 작성 문법 등 약간의 차이점은 존재
 

## 본 셸(Bourne shell)

 - 단순하고 처리 속도가 빠른 이유로 많이 사용
 - 시스템 관리 작업을 수행할 때 많은 스크립트는 본 셸을 기반으로 사용
 - 본 셸에서는 sh명령을 사용

## C 셸(C shell)

- 본 셸의 기능을 확장한 것
- 본 셸에 없던 에일리어스나 히스토리와 같은 사용자 편의 기능이 추가 제공
- 셸 스크립트의 작성의 문법 형식이 C언어와 비슷하여 이름을 갖게 됨
- C 셸은 다양한 편의 기능을 구현하기 위해 크기가 커지고 처리 속도가 느린 단점 등의 이유로
  편리한 배시 셸에 밀리고 있으며 C 셸에서는 csh명령을 사용

## 콘 셸(Korn shell)

- C 셸과 달리 본 셸과 같은 호환성 및 편의 기능을 제공
- 처리속도가 빠른 장점
- 우분투 리눅스에서는 별도로 콘 셸을 설치하여 사용할 수 있음
- ksh명령을 사용

## 배시 셸(bash shell)

- 배시 셸은 본 셸을 기반으로 개발된 셸
- 본 셸과의 호환성을 유지하면서 C셸, 콘 셸의 편리한 기능까지 제공하는 셸
- 배시 셸은 GPL(General Public License)을 따라는 공개 ㅅ소프트웨어
- 리눅스 기본 셸로 제공
- 우분투 리눅스도 배시 셸을 기본 셸로 사용
- 배시 셸에서는 bash 명령을 사용

## 대시 셸(dash shell)

- 대시 셸은 본 셸 기반으로 개발
- POSIX 표준을 준수하여 보다 작은 크기로 개발된 셸
- 기존의 셸에 비해 대시 셸은 부팅 시 셸 스크립트를 빠르게 실행시킴
- 파일 크기가 작고 신뢰성이 높아 우분투 6.10버전부터
  본 셸 대신 대시셸을 시스템 기본 셸로 사용
 


- 터미널 창에서 심볼릭 링크로 dash가 설정되어 있는것을 확인할 수 있음

## 셸 변수 설정

- 셸 변수는 사용자가 지정한 이름으로 설정할 수 있음
- 셸 변수는 =을 사용하여 선언할 때 변수 이름과 문자열 사이에 공백이 존재해서는 안됨
- 변수의 이름은 대/소문자를 엄격하게 구별
- 셸 변수를 설정하는 형식
    - 기능 : 사용하고자 하는 셸 변수 선언
    - 형식 : 변수 이름=문자열
    - 예시 : INSA=Hello
    

- 사용자가 지정하는 변수명을 선언하고 선언한 변수에 간단한 문자열ㅇ르 대입한 다음 변수를 
  출력하는 과정

### step01

- 셸 변수 INSA를 선언하고 변수에 문자열 Hello를 입력하기 위해 당므과 같이 명령을 선언

 - 변수명과 문자열 사이에 공백이 존재하면 안 됨

### step02

- 선언한 변수에 대해 echo 명령을 수행하게 되면 변수에 대입된 문자열이 출력됨
- 변수를 출력할 때는 반드시 변수명 앞에 $를 붙여서 사용


- 변수명 앞에 $를 선언하지 않은 경우 변수명이 아닌 문자열 취급

- 변수명 앞에 $를 선언한 경우 변수에 대입한 문자열 출력
