# 1. 셸 스크립트의 장단점

- 셸 스크립트는 셸이나 커맨드라인 인터프리터에서 수행하도록 작성되었음
- 운영체제를 위해 사용된 스크립트를 의미
- 단순한 도메인 고유 언어로 여겨지기도 하지만 
  셸 스크립트가 수행하는 일반적인 기능은 파일 이용, 프로그램 실행, 문자열 출력 등이 있음

## 셸 스크립트의 장점

- 다른 프로그래밍 언어에서 작성된 코드보다 훨씬 빠르게 처리됨
- 셸 스크립트는 따로 컴파일을 수행하는 단계가 없음 
- 기존보다 더 빠르게 코드 실행


## 셸 스크립트의 단점

- 실행되는 각 명령에 대한 잠재적으로 새로운 하부 프로세스가 수많이 필요하게 됨 
- 속도가 느려질수 있음

- 더 많고 복잡한 스크립트들은 셸 스크립트 언어의 자체적인 제한영역에서 실행
- 다양한 셸이 문제를 개선할 목적으로 고품질의 코드와 확장을 기록하기에는
  다소 힘들다는 단점도 존재

# 2. 셸 출력 명령

- 터미널 창에서 문자열을 출력하는 리눅스 배시 셸의 출력 명령에는 echo와 printf 명령이 있음


## 공통 셸 출력 명령 : echo

- echo 명령은 모든 셸 스크립트 프로그래밍에서 공통적으로 사용할수 있는 출력 명령

- echo
    - 기능 : 화면에 한 줄의 문자열을 출력
    - 형식 : echo [옵션] [문자열]
    - 옵션 : 
        - -n : 문자열 중간에 선언하게 되면 선언한 위치에서 줄 바꿈을 수행

### step01 

- echo 명령 우분투 문자열 출력

### step02

- echo 명령으로 문자열을 출력할 때 문자열 사이에 공백이 존재할 경우 문자열을 큰 따옴표로 묶어 줘야 함

### step03
- echo 명령과 줄 바꿈 문자 옵션 -n을 함께 선언하게 되면 줄 바꿈 문자 다음에
  오는 문자열은 다음 행으로 줄이 바뀌어 출력됨


- 옵션 지정으로 프롬프트 전에 문자열이 출력됨

## 출력형식 지정 명령 : printf

- 출력형식을 %d 지시자와 \n 줄 바꿈 문자 등 C언어에서 사용하는 출력함수인 
  printf() 함수의 형식을 지정하여 출력할 수 있습니다.
  
- printf
    - 기능 : 출력형식을 지정하여 문자열을 출력
    - 형식 : printf [옵션] [문자열]
    - 옵션 :
        - %d, \n, 등 c언어의 printf()함수 형식을 지정

### step01

- 터미널창에서 printf 명령으로 'SPACE'문자열을 출력하면 프롬프튼 앞에 문자열이 출력

### step02

- 줄 바꿈 문자 \n 을 문자열 안에 선언하게 되면 일ㄹ부 문자열을 다음 행으로 줄을 바꿔 출력

### step03

- 출력 변환 기호 %d를 문자열 안에 선언하여 정수와 일대일 대응관계로 출력할수 있음
