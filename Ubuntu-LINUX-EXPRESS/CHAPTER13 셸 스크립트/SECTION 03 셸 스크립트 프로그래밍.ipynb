# 1. 셸 스크립트 작성과 실행

- 셸 스크립트 프로그램을 작성하는 방법은 주로 gedit나 vi 에디터를 사용하여 작성
- 리눅스의 많은 부분이 셸 스크립트로 작성


 - 셸 스크립트 파일을 저장하기 위해 /sh_test.d 디렉터리를 생성
 - 예제 작성 모든 파일을 이 디렉터리에 저장
 - 셸 스크립트 파일임을 쉽게 구별할 수 있도록 확장자 .sh를 붙여서 셸 스크립트 파일명 지정

## 셸 스크립트 작성 후 sh 명령으로 실행하기

 - 현재 시스템에 접속 중인 사용자 이름과 호스트 이름을 출력하는 셸 스크립트를 작성 후 
 - 터미널 창에서 sh 명령으로 셸 스크립트 파일을 실행하는 과정을 살펴보자

### step01

- 셸 스크립트 파일을 저장하기 위해 /sh_test.d 디렉터리를 생성한 다음 
- 작업 디렉터리의 위치를 /sh_test.d 디렉터리로 이동

### step02

- 터미널 창에서 명령을 수행한 다음 gedit 창에서 코드를 입력 후 저장함

```sh

#! /bin/sh
// 특별한 형태의 주석(#!)으로 bash를 사용하는 의미(생략하면 안됨)

echo " Name Print "
// 문자열 출력 echo

echo ">> Connect Name" $USERNAME
// $USERNAME 현재 시스템에 접속된 사용자 이름을 출력하는 환경변수 선언 

exit 0
// 종료 코드(0은 성공 의미)
```

- 셸 스크립트 myname.sh 파일을 실행

## 디렉터리의 위치가 다를 경우 : 셸 스크립트 실행 오류

- 이와 같이 작업하는 디렉터리의 위치를 선정한 이유는 sh 명령이 셸 스크립트 파일이
  저장되어 있는 디렉터리에서 실행되도록하기 위함
  
 

- 이와 같은 문제를 해결하려면 셸 스크립트 파일을 저장한 디렉터리에서 
  sh 명령을 수행하거나 `chmod +x 스크립트 파일명`을 사용하여 
  스크립트 파일을 `실행 가능`속성으로 변경해 주면 됨

# 2. 변수 선언

 - C, C++, Java 등의 프로그래밍에서는 변수를 사용하기 전에 먼저 변수를 선언하지만,
 - 셸 스크립트는 변수를 사용하기 전에 먼저 변수를 선언하지 않습니다.
 
 - 변수에 처음에 값이 할당되면서 변수가 자동으로 생성되기 때문

 - 하지만 변수 선언 시 유의해야 할 사항이 있음
 
 - 셸에서 사용되는 변수는 모두 문자열(String)로 취급
 - 숫자도 문자열로 처리
 
 - 알파벳 대/소문자 구별이 엄격
 - 변수에 값을 대입할 때는 `=`좌우에 공백이 없어야 함


- var1 = Hi <- 오류 좌우 공백
- var2=good
- var3=Nice Day <- 오류 두개의 문자열인 경우 큰 따옴표로 묶어야함
- var4="Good morning"
- var5=88+95 <- 오류는 없지만 연산은 안 됨(모두 문자열로 취급)|

## 변수의 입력과 출력

- 변수 선언의 유의사항을 준수하여 키보드로 입력받은 값을 변수에 저장하여 출력하는 프로그램

### step01

- 터미널 창에서 현재 디렉터리의 위치를 /sh_test.d 디렉터리로 이동한 다음 키보드로
  입력 받은 값을 1변수에 저장하고 출력하기 위해 터미널 창에서 명령을 수행한 다음 
  gedit 창에서 코드를 작성 후 저장

```sh

#! /bin/sh

echo " Q. Input : "
read input_string
// 키보드로 입력한 값을 변수 input_string에 저장하는 read 명령

echo " A. Output : $input_string "
// 키보드로 입력한 값이 저장된 변수 출력

exit 0
```

### step02 

- 셸 스크립트 inout.sh 파일을 실행한 다음 문자열 Hello를 키보드로 입력후 엔터

## 숫자 계산

- 셸 스크립트는 모든 변수를 문자열로 처리하기 때문에 연산을 수행하기 위해서는 
  키워드 expr을 사용해야 함
  
- 연산식을 사용하고자 할 때는 역따옴표(빽틱)와 expr 공백없이 붙여서 사용

- 빽틱으로 expr 연산식과 같이 시작과 끝을 묶어줘야 함

#### 연산식에서 괄호를 사용하려면 괄호 앞에  \를 반드시 붙여줘야 함
#### 사직연산(+,-,`*`,/)중에서 예외적으로 곱셈 연산(`*`)앞에도 반드시 \를 붙여줘야 함

### step01

- 산출하여 출력하기 위해 터미널 창에서 명령을 수행한 다음 gedit창에서 코드를 입력 후 저장

## 파라미터 변수

- 파라미터(Parameter) 변수는 자신이 실행한 명령어의 파라미터(매개변수)를 의미

- 파라미터 변수는 $0, $1, $2, ...등과 같이 실행하는 명령의 부분을 
  각각 하나씩 변수로 지정
  
- 예를 들어 패키지를 설치하는 dnf -y install xinetd 명령을 실행한다고 가정
- 파라미터 변수를 다음 표와 같이 지정

    - 명령어       : dnf -y install xinetd
    - 파라미터변수 :  $0  $1   $2     $3
